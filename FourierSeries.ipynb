```python
from sympy import symbols, fourier_series, pi, lambdify
import matplotlib.pyplot as plt
import numpy as np

# Define the function
x = symbols('x')
f = x**3
L = 2
N = [2, 5, 7]

fig, axs = plt.subplots(1, 3, figsize=(14, 4))

for i in range(len(N)):
    # Calculate the Fourier series
    s = fourier_series(f, (x, -L, L))
    s = s.truncate(N[i])
    
    # Plot the Fourier series
    x_vals = np.linspace(-2, 2, 100)
    y_vals = [s.subs(x, x_val) for x_val in x_vals]
    axs[i].plot(x_vals, y_vals, label="Fourier")

    # plot the main func
    f_np = lambdify(x, f, modules=['numpy'])
    y_vals = f_np(x_vals)
    axs[i].plot(x_vals, y_vals, label="f(x)=x^3")
    axs[i].set_ylabel('f(x)')
    axs[i].set_title('Fourier series of x^3 up to the {} term'.format(N[i]))
    axs[i].legend()

fig.tight_layout()
plt.show()
```
