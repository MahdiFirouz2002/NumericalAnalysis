```python
# import libraries
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
```

#### 1) Approximation of sin function by Taylor series

```python
# define symbols and functions
x = sp.Symbol('x')
f = sp.sin(x)

# calculate taylor series
taylor_series = sp.series(f, n=6)

# Print the Taylor series
print("taylor series of sin(x) up to 6 terms:")
display(sp.Eq(sp.Symbol('sin(x)'), taylor_series))

# note: removeO used for remove order notation of calculated series 
taylor_series = taylor_series.removeO()

# Plot the Taylor series and the original function
x_vals = np.linspace(-np.pi, np.pi, 1000)
y_vals = [sp.sin(i) for i in x_vals]
taylor_vals = [taylor_series.subs(x, i) for i in x_vals]

plt.plot(x_vals, y_vals, label='sin(x)')
plt.plot(x_vals, taylor_vals, label='Taylor series')
# plt.legend for display plot labels
plt.legend()
plt.show()
```

#### 2) Approximation of cos function by Taylor series
in this example we try to decrease error of approximated function under $10^{-7}$

```python
# define symbols and functions
x = sp.Symbol('x')
f = sp.cos(x)
eps = 1e-7
max_error = 1
n = 1

while max_error > eps:
    # calculate taylor series
    taylor_series = sp.series(f, n=n)

    # Print the Taylor series
    print("taylor series of cos(x) up to {} terms:".format(n))
    display(sp.Eq(sp.Symbol('cos(x)'), taylor_series))

    # note: removeO used for remove order notation of calculated series 
    taylor_series = taylor_series.removeO()

    # Plot the Taylor series and the original function
    x_vals = np.linspace(-2, 2, 1000)
    y_vals = [np.cos(i) for i in x_vals]
    taylor_vals = [taylor_series.subs(x, i) for i in x_vals]

    # Calculate the error at each x value
    error_vals = [abs(f.subs(x, i) - taylor_series.subs(x, i)) for i in x_vals]
    max_error = max(error_vals)
    print("maximum Error: ", max_error)

    plt.plot(x_vals, y_vals, label='cos(x)')
    plt.plot(x_vals, taylor_vals, label='Taylor series')
    plt.plot(x_vals, error_vals, label='Error')
    plt.legend()
    plt.show()
    print("------------------------------------------------------------------")

    n += 1 
```
